```python
data['Departure Time'] = [i.split(':') for i in data['Departure Time']]
```

```python
lst = [[int(i[0]),int(i[1])] for i in data['Departure Time']]
```

```python
lst1 = []
for i in lst:
    if i[0]>12 and i[0]<20:
        a = i[0]-2
        b = str(a)
        print(b)
        lst1.append((int(b[1])*60)+i[1])
    elif i[0]>=20:
        a = i[0]-12
        b = (a*60)+i[1]
        lst1.append(b)
    else:
        lst1.append((i[0]*60)+i[1])
```

```python
data['Departure Time'] = lst1
```

```python
import re
arr_time = []
for i in data['Arrival Time']:
    a = re.findall(r'\d{2}:\d{2}',i)
    arr_time.append(a)
    
```

```python
arr_time1 = []
for i in arr_time:
    for j in i:
        arr_time1.append(j.split(':'))
```

```python
arr_time2 = [[int(i[0]),int(i[1])]for i in arr_time1]
```

```python
lst2 = []
for i in arr_time2:
    if i[0]>12 and i[0]<20:
        a = i[0]-2
        b = str(a)
        print(b)
        lst2.append((int(b[1])*60)+i[1])
    elif i[0]>=20:
        a = i[0]-12
        b = (a*60)+i[1]
        lst2.append(b)
    else:
        lst2.append((i[0]*60)+i[1])
```

```python
data['Arrival Time'] = lst2
```

```python
data.iloc[6474,7] = '5h'
```

```python
jr_time = []
for i in data['Journey Time']:
    a = re.findall(r'\d+',i)
    if len(a)==2:
        jr_time.append([int(a[0]),int(a[1])])
    else:
        jr_time.append([int(a[0])])
```

```python
jr_time1 = []
for i in jr_time:
    if len(i)>1:
        jr_time1.append((i[0]*60)+i[1])
    else:
        jr_time1.append(i[0]*60)
```

```python
data['Price'] = [int(i) for i in data['Price']]
```

```python
data.head()
```

```python
out = data['Price']
```

```python
inp = data.drop('Price',axis=1)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(inp,out,test_size=0.2,random_state=10)
```

```python
from catboost import CatBoostRegressor
from sklearn.metrics import mean_squared_error,r2_score,mean_squared_log_error
cb = CatBoostRegressor(depth=10, iterations =1200, learning_rate= 0.01,leaf_estimation_iterations=5)
cb.fit(xtrain,ytrain,cat_features=np.array([0,1,2,3,4,8,9]))
ypred = cb.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error,accuracy_score
mse = mean_squared_error(ytest,ypred)
mse**0.5
```
